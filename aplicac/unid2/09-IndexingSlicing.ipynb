```python
# Para ilustrar a indexação e o slicing, vamos usar a sequência de Fibonacci.

Fibonacci = [1,1]

while Fibonacci[-1] < 3000:
    Fibonacci.append(Fibonacci[-1]+Fibonacci[-2])

Fibonacci.pop()

Fibonacci
```

```python
# Para acessar uma entrada da lista, basta colocar a sua posição entre colchetes.
# Lembrando que a contagem das posições começa em zero.

Fibonacci[0]    # Obtenção do primeiro da lista
```

```python
Fibonacci[9]   # Obtenção do décimo da lista
```

```python
# Uma posição negativa será contada de trás para a frente, a partir do final da lista.

Fibonacci[-1]    # Obtenção do último da lista
```

```python
Fibonacci[-2]    # Obtenção do penúltimo da lista
```

```python
# Para acessar uma "fatia" da lista, basta usar dois pontos.
# O primeiro número é o índice do início da fatia.
# O segundo número é o índice da primeira entrada a ficar de fora da fatia.

Fibonacci[4:10]  # Obtenção do quinto ao décimo da lista
```

```python
# Se o início da fatia for omitido, a fatia começa no início da lista.

Fibonacci[:10]  # Obtenção de uma fatia até o décimo da lista
```

```python
# Se o final da fatia for omitido, a fatia termina no final da lista.

Fibonacci[4:]  # Obtenção da fatia, do quinto da lista em diante
```

```python
# Um terceiro parâmetro da "fatia" da lista pode ser fornecido para saltar elementos.
# Por exemplo, se passo for igual a 2, a fatia será retirada alternando os elementos.

Fibonacci[4:10:2]  # Obtenção do quinto ao décimo da lista
```
