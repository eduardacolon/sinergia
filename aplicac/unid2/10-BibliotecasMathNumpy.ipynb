```python
# A biblioteca NUMPY reúne mais ferramentas da Matemática, incluindo vetores e matrizes.

import numpy as np
```

```python
# Vetores podem ser declarados usando o tipo ARRAY

v = np.array([1,2,3,4,5])   # Declaração do vetor v, com 5 coordenadas

v                           # Exibição do vetor v
```

```python
# Vetor nulo

z = np.zeros(5)             # Declaração do vetor nulo com 5 coordenadas

z                           # Exibição do vetor z
```

```python
# Vetor de 1's

u = np.ones(5)             # Declaração do vetor de 1's com 5 coordenadas

u                           # Exibição do vetor u
```

```python
# Soma de vetores

u + v
```

```python
# Diferença de vetores

u - v
```

```python
# Multiplicação de vetor por escalar

5*u
```

```python
# Expressões algébricas com vetores

3*u - v
```

```python
# Produto escalar de dois vetores

np.dot(u,v)
```

```python
# A biblioteca NUMPY redefine todas as funções elementares, com uma implementação vetorial.
# Ao ser aplicadas a um vetor, as funções aplicam-se a cada coordenada do vetor.
# O resultado será um novo vetor, com o mesmo número de coordenadas.

np.exp(v)     # A função exponencial aplicada a cada coordenada de v
```

```python
# Além de vetores, o tipo ARRAY do NUMPY pode também acomodar matrizes.
# As matrizes podem ser declaradas linha por linha.

N = np.array([[1,2],[1,0],[3,2],[-1,0],[3,-2]])  # Exemplo de matriz com 5 linhas e 2 colunas

N
```

```python
# Matriz nula

Z23 = np.zeros([2,3])   # Matriz nula 2x3

Z23
```

```python
# Matriz identidade

I5 = np.identity(5)   # Matriz identidade 5x5

I5
```

```python
# Matriz de 1's

U45 = np.ones([4,5])   # Matriz de 1's 4x5

U45
```

```python
# Soma de matrizes do mesmo formato

N + N
```

```python
# Multiplicação de matriz por escalar

5*N
```

```python
# Multiplicação de matrizes de formato compatível

# Importante: a multiplicação de matrizes usa o símbolo de @, ao invés do *.

I5 @ N
```

```python
# Potência de matrizes quadradas

(2*I5)**3
```

```python
# O formato de uma matriz pode ser obtido pela propriedade SHAPE

N.shape
```
