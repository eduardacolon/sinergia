```python
# A biblioteca SciPy reúne as ferramentas matemáticas do Cálculo Avançado.

import numpy as np
import scipy.integrate as integrate  # Pacote de integração numérica
import scipy.optimize as optimize    # Pacote de otimização
import scipy.stats as stats          # Pacote estatístico
```

### Integração Numérica (SciPy)

```python
# Definição da função a ser integrada
def f(x):
    return np.exp(-x**2)
```

```python
# Cálculo da integral definida
I, erro = integrate.quad(f,0,1)

# Exibição do valor calculado e da margem de erro
print(f'Valor da integral: {I:.15f}')
print(f'Margem de erro: {erro:.15f}')
```

```python
# Cálculo da integral imprópria
I, erro = integrate.quad(f,0,np.inf)

# Exibição do valor calculado e da margem de erro
print(f'Valor da integral: {I:.15f}')
print(f'Margem de erro: {erro:.15f}')
```

### Otimização (SciPy)

```python
# Definição da função a ser usada na busca de raízes
def f(x):
    return x**3 - x
```

```python
# Busca de raízes da função
# Cada estimativa inicial pode levar a uma raiz diferente
# A convergência do método é indicada no campo "message"
# As raízes encontradas estão no vetor "x"
optimize.root(f,[-2,-0.2,5])
```

```python
# Lista de raízes, com arredondamento para 8 casas decimais
np.round(optimize.root(f,[-2,-0.2,5]).x,8)
```

```python
# Busca de pontos de mínimo local da função
optimize.minimize(f,1)
```

```python
# Ponto de mínimo local mais próximo da estimativa inicial 1
np.round(optimize.minimize(f,1).x[0],8)
```

```python
# Valor mínimo local no ponto encontrado
np.round(f(optimize.minimize(f,1).x[0]),8)
```

```python
# Busca de um ponto de máximo local da função

# Basta buscar um ponto de mínimo do oposto da função
def g(x):
    return -f(x)
optimize.minimize(g,-1)
```

```python
# Ponto de máximo local mais próximo da estimativa inicial 1
np.round(optimize.minimize(g,-1).x[0],8)
```

```python
# Valor máximo local no ponto encontrado
np.round(f(optimize.minimize(g,-1).x[0]),8)
```

### Estatística  (SciPy)
