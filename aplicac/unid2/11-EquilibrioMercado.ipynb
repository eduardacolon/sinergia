```python
# Carregamento da biblioteca NumPy.

import numpy as np
```

### Market share

Em uma cidade há 4 companhias de telefonia celular: A, B, C e D. O market share de cada companhia pode ser representado por um vetor de estado.

```python
# Lista das companhias telefônicas
Operadoras = ['A', 'B', 'C', 'D']

# O vetor de participação de mercado inicial
p = [np.array([0.40, 0.25, 0.20, 0.15])]

# Exibição da participação de mercado
print('Participação de mercado (market share):')
for k in range(len(Operadoras)):
    print(f'- Operadora {Operadoras[k]}: {p[0][k]:,.2%}')
```

### Mudança de operadora

A cada ano muitos clientes que trocam de operadora, isso pode ser representado por uma matriz de transição, indicando o percentual de clientes que troca de operadora.

Aqui, cada entrada da matriz tem sua linha correspondente à nova operadora e sua coluna, à operadora de origem, enquanto que seu valor mostra o percentual de clientes que deixam a operadora de origem e mudam para a nova operadora a cada ano. (Note que a diagonal principal indica os clientes que permanecem na mesma operadora.)

```python
# Definição da matriz de transição
T = np.array([[0.40, 0.20, 0.15, 0.10],[0.15, 0.40, 0.15, 0.20],[0.35,0.20, 0.50, 0.10],[0.10,0.20, 0.20, 0.60]])

# Exibição a matriz de transição
T
```

### Evolução do market share

Ao longo dos anos, à medida que os clientes mudam de operadora, as participações de mercado vão-se alterando.

```python
# Evolução da participação de mercado ao londo de N anos

N = 3  # Número de anos à frente

for k in range(N):
    p.append(T @ p[k])   # Cada passo inclui um novo ano, usando a matriz de transição
    
# Exibição daa evolução do market share
print('Evolução do market share:')

Cabec = 'Ano     '
for x in Operadoras:
    Cabec += f'{x:8s}'
print(Cabec)

for k in range(N+1):
    Linha = f'{k:2}    '
    for j in range(len(Operadoras)):
        Linha += f'{p[k][j]:,.2%}  '
    print(Linha)
```

### Equilíbrio de mercado

A grande questão é: será que existe um equilíbrio de mercado? Ou seja, será que, ao longo do tempo as participações de mercado de cada operadora tendem a se estabilizar?

Note que, se houver um equilíbrio de mercado, ele corresponderá a um autovetor da matriz de transição, correspondente ao autovalor 1. Por isso, matematicamente, a busca de um equilíbrio de mercado corresponde à busca de autovetores do autovalor 1 da matriz.

```python
# Cálculo de autovalores e autovetores
autovalores, autovetores = np.linalg.eig(T)
```

```python
# Exibição da lista de autovalores
autovalores
```

```python
# Exibição da matriz de autovetores
# Cada coluna é um autovetor correspondente aos autovalores encontrados
autovetores
```

```python
if 1 in np.round(autovalores,8):   # Teste de existência de autovalor 1
    print('Equilíbrio de mercado:')
    # Seleção da posição do autovalor 1 na lista de autovalores
    AutoVal1pos = np.where(np.round(autovalores,8) == 1)[0][0]
    # Seleção da coluna da matriz de autovetores, correspondente ao autovalor 1
    AutoVet1 = autovetores[:,AutoVal1pos]     
    # Cálculo do ponto de equilíbrio
    Equilibrio = AutoVet1/AutoVet1.sum()
    # Exibição do equilíbrio de mercado
    for k in range(len(Operadoras)):
        print(f'- Operadora {Operadoras[k]}: {Equilibrio[k]:,.2%}')
        
else:
    print('Não existe equilíbrio de mercado.')
```
