```python
# Carregamento das bibliotecas
import numpy as np
import pandas as pd
import scipy.optimize as optimize    # Pacote de otimização
```

### Parte I - Cálculo de juros

#### Cálculo do montante

Uma dívida com principal $P$ foi contraída a uma taxa de juros mensal $r$, com um prazo de $t$ meses para pagamento.
Qual o montante dessa dívida ao final do prazo dado?

```python
P = 5000.00  # Principal (em R$)
r = 0.023    # Taxa de juros (em % a.m.)
t = 18       # Tempo (em meses)

# Cálculo do montante, usando juros compostos
M = P*(1+r)**t

# Exibição do resultado
print('Cálculo do montante:')
print(f'- Principal:     R$ {P:,.2f}')
print(f'- Taxa de juros: {100*r:,.2f}% a.m.')
print(f'- Prazo:         {t} meses')
print(f'=> Montante:     R$ {M:,.2f}')
```

#### Desconto racional (por dentro)

Um montante $M$ é devido para pagamento daqui $t$ meses. Caso se deseje antecipar o pagamento, qual seria o valor presente usando uma taxa de juros $r$ para fazer o desconto por dentro?

```python
M = 5000.00  # Montante (em R$)
r = 0.023    # Taxa de juros (em % a.m.)
t = 3        # Tempo (em meses)

# Cálculo do principal, usando juros compostos para fazer o desconto
P = M/((1+r)**t)

# Exibição do resultado
print('Cálculo do valor presente (desconto por dentro):')
print(f'- Valor futuro:  R$ {M:,.2f}')
print(f'- Taxa de juros: {100*r:,.2f}% a.m.')
print(f'- Prazo:         {t} meses')
print(f'=> Valor presente: R$ {P:,.2f}')
```

#### Desconto comercial (por fora)

Um montante $M$ é devido para pagamento daqui $t$ meses. Caso se deseje antecipar o pagamento, qual seria o valor presente usando uma taxa de juros $r$ para fazer o desconto por fora?

```python
M = 5000.00  # Montante (em R$)
r = 0.023    # Taxa de juros (em % a.m.)
t = 3        # Tempo (em meses)

# Cálculo do principal, usando juros simples para fazer o desconto
P = M*(1-r*t)

# Exibição do resultado
print('Cálculo do valor presente (desconto por fora):')
print(f'- Valor futuro:  R$ {M:,.2f}')
print(f'- Taxa de juros: {100*r:,.2f}% a.m.')
print(f'- Prazo:         {t} meses')
print(f'=> Valor presente: R$ {P:,.2f}')
```

#### Cálculo da taxa de juros

Um principal $P$ foi emprestado por um prazo de $t$ meses, resultando em um montante $M$. Qual é a taxa de juros usada nessa operação?

```python
P = 4655.00  # Principal (em R$)
M = 5000.00  # Montante (em R$)
t = 3        # Tempo (em meses)

# Cálculo da taxa de juros, usando juros compostos
r = (M/P)**(1/t) - 1

# Exibição do resultado
print('Cálculo da taxa de juros:')
print(f'- Principal: R$ {P:,.2f}')
print(f'- Montante:  R$ {M:,.2f}')
print(f'- Prazo:     {t} meses')
print(f'=> Taxa de juros: {100*r:,.2f}% a.m.')
```

#### Cálculo do prazo

Quanto tempo leva para que um principal $P$, emprestado a uma taxa de juros $r$, resulte em um montante $M$?

```python
P = 5000.00   # Principal (em R$)
M = 10000.00  # Montante (em R$)
r = 0.03      # Taxa de juros (em % a.m.)

# Cálculo do prazo, usando juros compostos
t = np.log(M/P)/np.log(1+r)

# Exibição do resultado
print('Cálculo do prazo:')
print(f'- Principal:     R$ {P:,.2f}')
print(f'- Montante:      R$ {M:,.2f}')
print(f'- Taxa de juros: {100*r:,.2f}% a.m.')
print(f'=> Prazo:        {t:.1f} meses')
```

### Parte II - Cálculos para financiamentos (Tabela Price)

#### Cálculo da prestação

Qual será o valor da prestação mensal para financiar um valor $P$ a uma taxa de juros $r$ em $n$ pagamentos iguais?

```python
P = 50000.00 # Valor a ser financiado (principal)
r = 0.02     # Taxa de juros (em % a.m.)
n = 18       # Número de prestações mensais (prazo)

# Cálculo da prestação, usando a Tabela Price
PGTO = P * r / (1 - (1+r)**(-n))

# Exibição do resultado
print('Cálculo do valor da prestação:')
print(f'- Valor a ser financiado:       R$ {P:,.2f}')
print(f'- Taxa de juros:                {100*r:,.2f}% a.m.')
print(f'- Número de prestações mensais: {n}')
print(f'=> Valor da prestação: R$ {PGTO:,.2f}')
```

#### Cálculo da taxa de juros implícita

Um valor $P$ pode ser financiado em $n$ prestações mensais de valor fixo $PGTO$. Qual é a taxa de juros desse financiamento?

```python
P = 2800.00   # Valor a ser financiado (principal)
n = 20        # Número de prestações mensais (prazo)
PGTO = 200.00 # Valor da prestação mensal

# Cálculo da taxa de juros, usando a Tabela Price

# Definição do valor da prestação como função da taxa de juros
def f (x):
    return P * x / (1 - (1+x)**(-n)) - PGTO

r = optimize.root(f,[0.001,1]).x[0]

# Exibição do resultado
print('Cálculo da taxa de juros implícita:')
print(f'- Valor a ser financiado:       R$ {P:,.2f}')
print(f'- Valor da prestação mensal:    R$ {PGTO:,.2f}')
print(f'- Número de prestações mensais: {n}')
print(f'=> Taxa de juros: {100*r:,.2f}% a.m.')
```

### Parte III - Cálculos para séries de fluxos de caixa

#### Cálculo do valor presente líquido

Qual é o valor presente líquido de uma série de fluxos de caixa $F$, descontada por uma taxa de juros $r$?

OBS: Para o cálculo do VPL, deve-se usar sempre o desconto racional de cada fluxo de caixa.

```python
F = [100.00,100.00,100.00] # Série de fluxos de caixa mensais
r = 0.05                   # Taxa de juros (em % a.m.)

# Uma variável VPL será usada para acumular os valores presentes de cada fluxo de caixa da série
VPL = 0

# Cálculo do VPL
for t in range(len(F)):
    VPL += F[t]/((1+r)**t)   # Soma o valor presente de cada fluxo de caixa da série

# Exibição do resultado
print('Cálculo do valor presente líquido de uma série de fluxos de caixa:')
print(f'- Série de fluxos de caixa:')
for t in range(len(F)):
    print(f'  F[{t}] = R$ {F[t]:,.2f}')
print(f'- Taxa de juros: {100*r:,.2f}% a.m.')
print(f'=> VPL: R$ {VPL:,.2f}')
```

#### Atualização de um valor pelo IPCA

Qual seria o valor atualizado de uma nota de R$ 100 em 1º de julho de 1994? 
Valor atualizado para 1º de abril de 2023 pelo IPCA mensal.

```python
# Carrega os dados do IPCA mensal de um arquivo Excel, de julho de 1994 a março de 2023
IPCA = pd.read_excel('IPCA.xlsx')
```

```python
PV = 100.00   # Valor nominal no início do plano real

# Variável para o valor atualizado para abril de 2023
FV = PV       # Inicialmente ajustada para o valor nominal original

# Atualização mensal para todos os meses do período
for k in range(len(IPCA)):
    FV *= 1 + IPCA['IPCA'][k]/100      # Corrige pelo IPCA do mês
        
# Exibição do resultado
print (f'Um valor nominal de R$ {PV:,.2f} em julho de 1994 corresponde a R$ {FV:,.2f} em abril de 2023.')
```
