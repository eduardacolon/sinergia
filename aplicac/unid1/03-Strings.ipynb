```python
a = 'Olá, '               # Uma variável textual.
b = 'José'                # Outra variável textual.
a + b + '!'               # As duas variáveis e um texto com um ponto de exclamação concatenados
```

```python
# Um dado do tipo string pode ser multiplicado por um número inteiro, produzindo a sua repetição.
15*a + b +'!'
```

```python
n = 12      # Uma variável numérica.
```

```python
# A variável numérica não pode ser diretametne concatenada com strings.
'O ano tem '+ n +' meses.'
```

```python
# Porém, a variável numérica pode ser convertida para o tipo string e concatenada.
'O ano tem '+ str(n) +' meses.'
```


Observação sobre a conversão de tipos de dados.

```python
# Exibindo a variável como ela foi declara
n
```

```python
# Note que se trata de um tipo inteiro
type(n)
```

```python
# Convertendo a variável para número real
float(n)
```

```python
# Convertendo a variável para número complexo
complex(n)
```

```python
# Convertendo a variável para texto
str(n)
```

```python
# Note que o tipo da variável não foi alterado!
# As conversões acima só foram feitas para exibição e não atribuídas à variável...
type(n)
```

```python
# Para alterar o valor ou o tipo de uma variável, ela tem que sofrer uma nova atribuição.
n = complex(n)
```
