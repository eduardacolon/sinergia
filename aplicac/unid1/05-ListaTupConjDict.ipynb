### Dados lógicos

```python
# Dados lógicos podem assumir apenas dois valores: verdadeiro (True) ou falso (False).
a = False
b = True
```

```python
# Operador de negação - NÃO.
not a
```

```python
# Operador de conjunção - E.
a and b
```

```python
# Expressão lógica combinando tipos e operadores lógicos.
(not a) and b
```

```python
# Operador de disjunção - OU.
a or b
```

```python
# Expressão lógica combinando tipos e operadores lógicos.
a or (not b)
```

### Coleções de dados

```python
# Lista de dados numéricos, definida com colchetes.
# Um mesmo valor pode aparecer repetidas vezes.
a = [1,1,2,3,5,8]
```

```python
# Lista de dados textuais, definida com colchetes.
a = ['José','Maria']
```

```python
# A função len retorna o número de entradas de uma lista.
len(a)
```

```python
# Os elementos de uma lista podem ser acessados por sua posição na lista, começando de 0.
a[1]
```

```python
# A lista pode ser exibida digitando seu nome.
a
```

```python
# Ênuplo de dados numéricos, definido com parênteses.
b = (1,3,4)
```

```python
# A tupla pode ser exibida digitando seu nome.
b
```

```python
# A diferença entre listas e tuplas é que as entradas de uma tupla não podem ser alteradas.
b[0]=2
```

```python
# Conjunto de dados numéricos, definido com chaves.
c={1,1,2,3,5,8}
```

```python
# O conjunto pode ser exibido digitando seu nome.
# Note que valores repetidos são ignorados e a ordem não é levada em consideração.
c
```

```python
# Não é possível acessar os elementos de um conjunto por sua posição.
c[0]
```

```python
# A relação entre um conjunto e seus elementos é a relação de pertinência
# A palavra-chave in testa se um elemento pertence ao conjunto
# Quando um elemento pertence ao conjunto o resultado do operador in é verdadeiro
1 in c
```

```python
# Quando um elemento não pertence ao conjunto o resultado do operador in é falso
4 in c
```

```python
# Dicionário de dados numéricos, definido com chaves, com elementos indexados por texto.
d = {'a':1,'b':2,'c':3,'d':5,'e':8}
```

```python
# O dicionário pode ser exibido digitando seu nome.
d
```

```python
# Os elementos de um dicionário podem ser acessados por sua chave.
d['a']
```
