```python
# Exemplo do uso da estrutura iterativa FOR
# Exibição dos nomes de uma lista de chamada

a = ["Maria","José","Antônia","Pedro","Maria"]   # Nomes dos alunos de uma turma

for x in a:     # O código a seguir será executado para cada entrada da lista
    print(f'Nome: {x}')
```

```python
# Observe que, se fosse usado um conjunto em vez de uma lista, o resultado seria diferente
# Isso ocorre porque conjuntos não levam em consideração nem ordem, nem valores repetidos

b = {"Maria","José","Antônia","Pedro","Maria"}   # Nomes dos alunos de uma turma

for x in b:     # O código a seguir será executado para cada entrada da lista
    print(f'Nome: {x}')
```

```python
# Exemplo do uso da estrutura iterativa FOR
# Divisão de uma lista de números inteiros por 2, considerando a divisão inteira com resto

c = [0,1,2,3]

for n in c:
    print(f'Número {n} divido por 2: {n//2} com resto {n%2}')
```

```python
# Exemplo do uso da estrutura iterativa FOR, com o comando RANGE
# Divisão de uma lista de números inteiros por 2, considerando a divisão inteira com resto

# Aqui o comando RANGE é usado para substituir a lista de inteiros de 0 até 4, excluindo o 4

for n in range(4):
    print(f'Número {n} divido por 2: {n//2} com resto {n%2}')
```

```python
# Exemplo do uso da estrutura iterativa FOR, com o comando RANGE
# Divisão de uma lista de números inteiros por 2, considerando a divisão inteira com resto

# O comando RANGE gera uma progressão artimética (P.A.) e tem 3 parâmetros:
# - Start: o primeiro termo da P.A. (se não for declarado, esse valor é considerado 0)
# - Stop: o valor no qual a P.A. será encerrada (excluindo esse valor)
# - Step: a razão da P.A. (se naõ for declarado, esse valor é considerado 1)

for n in range(1,12,3):
    print(f'Número {n} divido por 2: {n//2} com resto {n%2}')
```

```python
# Aqui consideramos a sequência dos quadrados dos primeiros números naturais

s = [1,4,9,16]   # Definição da sequência completa

s                # Exibição da sequência
```

```python
# A mesma sequência poderia ser gerada usando um comando FOR

s = []          # Definição de uma sequência vazia

for n in range(1,5):     # Loop com 4 passos, de 1 a 4
    s.append(n**2)       # Inclusão de cada novo quadrado na sequência

s                # Exibição da sequência
```

```python
# A grande vantagem de usar o FOR é que os termos da sequência são preenchidos automaticamente.
# Assim, é muito mais fácil de gerar sequências com muitos termos.

for n in range(5,11):    # Loop com 6 passos, de 5 a 10, continuando a sequência anterior
    s.append(n**2)       # Inclusão de cada novo quadrado na sequência

s                # Exibição da sequência
```

```python
# Outra forma de gerar a mesma sequência, usando um comando FOR dentro da definição da lista

s = [n**2 for n in range(1,11)]    # Definição da sequência completa

s                                 # Exibição da sequência
```

```python
# Exemplo de sequência gerada por recursividade, com o comando FOR.
# A sequência de Fibonacci tem os dois primeiros termos iguais a 1.
# A partir do terceiro termo, cada termo é igual à soma dos dois termos anteriores.

Fibonacci = [1,1]           # Definição dos dois termos iniciais.

for n in range(10):        # Geração recusriva de mais 10 termos da sequência de Fibonacci
    Fibonacci.append(Fibonacci[-1]+Fibonacci[-2])
    
Fibonacci
```

```python
# Exemplo de sequência gerada por recursividade, com o comando WHILE.
# A sequência de Fibonacci pode ser gerada de forma alternativa.

Fibonacci = [1,1]

while Fibonacci[-1] < 1000:
    Fibonacci.append(Fibonacci[-1]+Fibonacci[-2])

Fibonacci.pop()

Fibonacci
```
