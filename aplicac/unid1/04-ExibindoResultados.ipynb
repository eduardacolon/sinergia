```python
P = 5000.00  # Principal (em R$)
r = 0.023    # Taxa de juros (em % a.m.)
t = 18       # Tempo (em meses)

# Cálculo do montante, usando juros compostos
M = P*(1+r)**t
```

```python
# Exibição das variáveis
P, r, t, M
```

```python
# Montante arredondado para 2 casas decimais
round(M,2)
```

```python
# Resumo dos cálculos realizados em formato de texto

print('Cálculo do montante de um investimento')
print('(Juros compostos)')
print()
print(f'Principal:     R$ {P:,.2f}')
print(f'Taxa de juros: {r:.2%}a.m.')
print(f'Tempo:         {t} meses')
print()
print(f'Montante:      R$ {M:,.2f}')
```
