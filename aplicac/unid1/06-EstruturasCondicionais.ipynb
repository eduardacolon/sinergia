```python
# Exemplo de uso da estrutura condicional IF-ELSE
# Divisão de dois números, testando se o divisor não é zero

a = 4     # Dividendo
b = 3     # Divisor

if b != 0:        # Testa se o divisor for não nulo
    q = a / b                                   # Executa a divisão
    print (f'O resultado da conta é {q}')       # Mostra o resultado
else:             # Código a executar se o divisor for nulo
    print ('A divisão por zero não é possível.')
```

```python
# Exemplo de uso da estrutura condicional IF-ELIF-ELSE
# Atribuição de um conceito a partir da nota em uma disciplina

Nota = 92   # Nota na disciplina

print(f'Nota: {Nota}')     # Exibe a nota obtida
    
if Nota >= 90:   # Teste para o conceito A
    print('Conceito: A')
elif Nota >= 80: # Teste para o conceito B
    print('Conceito: B')
elif Nota >= 70: # Teste para o conceito C
    print('Conceito: C')
elif Nota >= 60: # Teste para o conceito D
    print('Conceito: D')
else:            # Naso residual, atribui o conceito F
    print('Conceito: F')
    
```
